```python
lds, lgt = create_dataset_line(400, m_s=[3, -0.5, 1.5, -0.9], centers=[(-1, -3), (-2, 2), (-5, 0), (-3,-5)], noise_var=0.2)

write_np_array(name="lines_rects4.csv", path=joinpath("..", "..", "datasets", "2d", "lines", "no_outliers"), x=normalize_points(lds))
write_np_array(name="lines_rects4_gt.csv", path=joinpath("..", "..", "datasets", "2d", "lines", "no_outliers"), x=lgt, dimensions=1)

outliers = np.hstack((np.random.uniform(-7.5,1.5, size=(len(lds), 1)), np.random.uniform(-9, 4, size=(len(lds), 1))))
lds = np.vstack((lds, outliers))
lgt = np.concatenate((lgt, np.zeros(shape=(len(outliers)), dtype=int)))
lds = normalize_points(lds)

plot_clusters(lgt, lds)
```

```python
pds, pgt = create_dataset_parabola(300, a_s=[1, -1.5, -0.9], centers=[(2, -4), (-7, -2), (-3,-5)], rot_angle=[-20, 180, 0], noise_var=0.2)

write_np_array(name="circles_parable3.csv", path=joinpath("..", "..", "datasets", "2d", "circles", "no_outliers"), x=normalize_points(pds))
write_np_array(name="circles_parable3_gt.csv", path=joinpath("..", "..", "datasets", "2d", "circles", "no_outliers"), x=pgt, dimensions=1)

outliers = np.hstack((np.random.uniform(-7.5,2.5, size=(len(pds), 1)), np.random.uniform(-10,4, size=(len(pds), 1))))
pds = np.vstack((pds, outliers))
pgt = np.concatenate((pgt, np.zeros(shape=(len(outliers)), dtype=int)))
pds = normalize_points(pds)

plot_clusters(pgt, pds)
```
