```python
def compute_implicit_form(data):
    (x1, y1), (x2, y2) = data
    # x1, y1 = p1
    # x2, y2 = p2
    a = y2 - y1
    b = x1 - x2
    c = x2*y1 - y2*x1
    return a, b, c
```

```python
def compute_dist(pars, p):
    a, b, c = pars
    return (a*p[:,0] + b*p[:,1] + c) / np.sqrt(a**2 + b**2)
```

```python
line = LineEstimator()
for _ in range(1000):
    ps = np.random.uniform(-10, 10, size=(2, 2))
    line.fit(ps)
    sl1, it1 = round(line.slope, 5), round(line.intercept, 5)
    a, b, c = compute_implicit_form(ps)
    sl, it = round(-a/b, 5), round(-c/b, 5)
    d1 = round(line.get_residuals(ps[0].reshape(1, -1))[0], 5)
    d = round(compute_dist((a, b, c), ps[0].reshape(1, -1))[0], 5)
    print(sl1 == sl, it1 == it, d==d1)
```
